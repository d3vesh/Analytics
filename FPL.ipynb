## Calculating team ROI
We plotted each Team’s cumulative player ROI against the number of players that the coach uses on a regular basis (players who have played at least 360 minutes so far in the season). This will help us identify the teams that have too many expensive and under-performing players who rarely play the full 90 minutes each game due to the frequent squad rotation their coach employs, which makes them a bad investment in the long run, since they will not be generating fantasy points consistently each game. Also, this graph below will help us identify the teams with coaches that do not use frequent player rotation, which results in those teams having a more consistent core of regular squad players. That will inform our algorithm to pick more players from those teams since their players are expected to generate a higher aggregate ROI in the long run because they would be involved in a lot more game action on average compared to players from teams that use frequent player rotation.



ROI is defined as (Team's Fantasy Points)/ (Team's Fantasy Cost) in other words total points returned per 1 million fantasy money spent.

# Calculating the Factor
Here we'll calculate the factor which will account for the combined effect of Fixture Difficulty Ratio(FDR) and Home/ Away game. The information related to this is obtained from FPL_factor dataframe we imported in the importing and reading section.


```python
FPL_factor.iloc[0]
```

This factor is calculated by using weighted average method where, FDR of every game week is multiplied by a weight depending on the game whether it is Home game or away game. <br> If it is a home game then the weight id "0.8" and "1.2" for away games, that is if a team has home game then it will reduce the effect of FDR and if the team has away game then it will increase the effect of FDR. <br>
For example, if the FDR of a team for the next game week is 4 and it's an home game then factor for the week is reduced as 0.8 gets multiplied, logically a home game against a strong team (high FDR) will be easier then a away game against that team.

```python
factor = [ ]
denom = 0
num = 0
for index in range(20):
    for column in range(2,16,3):
        fdr = int(FPL_factor.iloc[index,column])
        
        if FPL_factor.iloc[index,(column+1)] == "H":
            ha = 0.8
        elif FPL_factor.iloc[index,(column+1)] == "A":
            ha = 1.2
        multi = fdr*ha
        denom += ha
        num += multi
    factor.append("%.2f" %(num/denom))
        
print(factor)
```

```python
# Creating Factor dataframe
FDR_HA_factor = pd.DataFrame({"Team":FPL_factor.Club,"Factor":factor})
print(FDR_HA_factor)
```

# Python Algorithm to get most Optimum Player

```python
def my_team (budget = 100, star_player_limit = 3, gk = 2, df = 5, mid = 5, fwd = 3 ): # Pass constraints to function
    team = [ ]                                      # List of team to be returned
    star_position = [ ]                             # list containing position of starplayer
    star_player_limit = star_player_limit
    budget = budget
    injured = dataset2.loc[(dataset2.loc[:,"Status"] == 'injured'),"Name"] # Keeping a check of injury status
    positions = {"GKP":gk,"DEF":df,"MID":mid,"FWD":fwd}       # Dict accounting for no. of postions left to fill
 # Selecting top point player (Star Player)
    for ind in Top_points.index:       # Looping through the dataframe of players
        player = Top_points.loc[ind]   # Row of Dataframe one at a time
        if (len(team) < star_player_limit and player.Name not in injured and budget > player.Cost and positions[player.Position] > 0 and player.Position not in star_position ):
            team.append(player)
            budget -= player.Cost
            positions[player.Position] -= 1
            star_position.append(player.Position)    # Checking position of star player
 # Selecting normal player
    for ind in normal_player.index:
        player = normal_player.loc[ind]
        if (player.Name not in (x[0] for x in team) and player.Name not in injured and budget > player.Cost and positions[player.Position] > 0 ):
            team.append(player)
            budget -= player.Cost
            positions[player.Position] -= 1
    Squad = pd.DataFrame(team).sort_values(by = "Factor",ascending = True)    # Entire Squad of 15 player
    # Considering Factor for selecting playing 11
    GK = Squad.loc[(Squad.loc[:,"Position"] ==  "GKP"),:].sort_values(by = "Factor", ascending =  True).head(1)
    Others = Squad.loc[(Squad.loc[:,"Position"] != "GKP"),:].sort_values(by = "Factor", ascending =  True).head(10)
    Playing_11 = Others.append(GK).reset_index(drop = True)
    return (Squad,Playing_11)
```
